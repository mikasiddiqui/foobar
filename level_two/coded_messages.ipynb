```python
from itertools import permutations
```

```python
test_in = [3, 1, 4, 1, 5, 9]
test_in = list(test_in)
perm = [int(''.join([str(x) for x in l])) for i in range(len(test_in)) for l in permutations(test_in, i+1)]
```

```python
perm
```

```python
max_curr_result = 0
for num in perm:
    if  =num % 3= 0 and num > max_curr_result:
        max_curr_result = num
```
